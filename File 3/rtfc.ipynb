```python
from cryptography.fernet import Fernet
key = Fernet.generate_key()
f = Fernet(key)
with open('my_key','wb') as fail:
    fail.write(key)
```

```python
def load_key():
    return open('my_key','rb').read()
my_key = load_key()
fr = Fernet(my_key)
import bytes
for u in dat['Postal Code']:
    df = str(u).encode('utf-8')
    enc = fr.encrypt(df)
    with open('help','ab') as fail:
        fail.write(enc+b'\n')
```

```python
my_key = load_key()
fr = Fernet(my_key)
with open('dishf','ab') as an:
    with open('help','rb') as fail:
        for i in fail:
            dishf = fr.decrypt(i)
            an.write(dishf+b'\n')
```
